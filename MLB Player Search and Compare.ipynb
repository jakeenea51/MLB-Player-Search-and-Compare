```python
# Necessary imports

import math
import pandas as pd
from tkinter import *
from tkinter import messagebox

#quick startup
btc = pd.read_csv('MLB-stats-2021-b.csv') #complete season for batters
ptc = pd.read_csv('MLB-stats-2021-p.csv') #complete season for pitchers 

#store player index values
bi = pd.Series(btc['Name'], index = range(0,1508)) #batter index
pi = pd.Series(ptc['Name'], index = range(0,909)) #pitcher index
```

```python
root = Tk()
root.title('MLB 2021 Player Search and Compare')
root.geometry('810x150')
root.config(bg='white')
root.iconbitmap('images/mlb-icon.ico')

#list of batter stats
stats = btc.columns.tolist()
#list of pitcher stats
pstats = ptc.columns.tolist()

#variable for current row
global currentRow
currentRow=4

#variable for fontSize
fontSize = 8


#search function
#determines whether to follow batter function or pitcher function
def search(name):
    global playerType
    if playerType(name) == 'b':
        batterSearch(name)
    elif playerType(name) == 'p':
        pitcherSearch(name)
    else:
        response = messagebox.showerror('Invalid Player', "Player does not exist. Try re-typing the name of the player you are looking for. (Case sensitive)")
        
        
        
        
#playerType function
#determines player type
def playerType(name):
    if name in ptc['Name'].tolist() and btc['Name'].tolist():
        pa = btc.loc[list(bi).index(name)].at['PA']
        ip = ptc.loc[list(pi).index(name)].at['IP']
        if pa > ip:
            return 'b'
        else:
            return 'p'
    else:
        if name in ptc['Name'].tolist():
            return 'p'
        elif name in btc['Name'].tolist():
            return 'b'
        else:
            return 'n'

    
    
    
    
#compare function
#displays a second player's stats and compares the two  
def compare(name):
    global playerType
    firstPlayer = searchBar.get()
    
    #compare batters
    if playerType(name) == 'b' and playerType(firstPlayer) == 'b':
        Label(root, text= str(btc.loc[list(bi).index(name)].at['Name']), width=32, relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+2, column=0, padx=(10,0))
        Label(root, text= str(btc.loc[list(bi).index(name)].at['Team']), width=6, relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+2, column=1)
                                                                                                                                
        
        #compare stats and color boxes accordingly
        j = 2
        for i in stats[2:]:
            secondPlayerStat = btc.loc[list(bi).index(name)].at[stats[j]]
            firstPlayerStat = btc.loc[list(bi).index(firstPlayer)].at[stats[j]]
            if stats[j] in ['SO', 'GDP', 'CS']:
                if secondPlayerStat < firstPlayerStat:
                    Label(root, text= str(secondPlayerStat), width=6, bg='green', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+2, column=j)
                    Label(root, text= str(firstPlayerStat), width=6, bg='red', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+1, column=j)
                elif secondPlayerStat > firstPlayerStat:
                    Label(root, text= str(secondPlayerStat), width=6, bg='red', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+2, column=j)
                    Label(root, text= str(firstPlayerStat), width=6, bg='green', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+1, column=j)
                else:
                    Label(root, text= str(secondPlayerStat), width=6, bg='yellow', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+2, column=j)
                    Label(root, text= str(firstPlayerStat), width=6, bg='yellow', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+1, column=j)
            else:
                if secondPlayerStat > firstPlayerStat:
                    Label(root, text= str(secondPlayerStat), width=6, bg='green', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+2, column=j)
                    Label(root, text= str(firstPlayerStat), width=6, bg='red', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+1, column=j)
                elif secondPlayerStat < firstPlayerStat:
                    Label(root, text= str(secondPlayerStat), width=6, bg='red', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+2, column=j)
                    Label(root, text= str(firstPlayerStat), width=6, bg='green', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+1, column=j)
                else:
                    Label(root, text= str(secondPlayerStat), width=6, bg='yellow', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+2, column=j)
                    Label(root, text= str(firstPlayerStat), width=6, bg='yellow', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+1, column=j)

            j = j+1
    
    #compare pitchers
    elif playerType(name) == 'p' and playerType(firstPlayer) == 'p':
        Label(root, text= str(ptc.loc[list(pi).index(name)].at['Name']), width=32, relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+2, column=0, padx=(10,0))
        Label(root, text= str(ptc.loc[list(pi).index(name)].at['Team']), width=6, relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+2, column=1)
        j = 2
        for i in pstats[2:]:
            secondPlayerStat = ptc.loc[list(pi).index(name)].at[pstats[j]]
            firstPlayerStat = ptc.loc[list(pi).index(firstPlayer)].at[pstats[j]]
            if pstats[j] in ['L', 'BS', 'IBB', 'BB', 'BK', 'WP', 'ERA', 'H', 'R', 'ER', 'HR', 'HBP']:
                if secondPlayerStat < firstPlayerStat:
                    Label(root, text= str(secondPlayerStat), width=6, bg='green', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+2, column=j)
                    Label(root, text= str(firstPlayerStat), width=6, bg='red', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+1, column=j)
                elif secondPlayerStat > firstPlayerStat:
                    Label(root, text= str(secondPlayerStat), width=6, bg='red', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+2, column=j)
                    Label(root, text= str(firstPlayerStat), width=6, bg='green', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+1, column=j)
                else:
                    Label(root, text= str(secondPlayerStat), width=6, bg='yellow', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+2, column=j)
                    Label(root, text= str(firstPlayerStat), width=6, bg='yellow', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+1, column=j)
            else:
                if secondPlayerStat > firstPlayerStat:
                    Label(root, text= str(secondPlayerStat), width=6, bg='green', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+2, column=j)
                    Label(root, text= str(firstPlayerStat), width=6, bg='red', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+1, column=j)
                elif secondPlayerStat < firstPlayerStat:
                    Label(root, text= str(secondPlayerStat), width=6, bg='red', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+2, column=j)
                    Label(root, text= str(firstPlayerStat), width=6, bg='green', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+1, column=j)
                else:
                    Label(root, text= str(secondPlayerStat), width=6, bg='yellow', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+2, column=j)
                    Label(root, text= str(firstPlayerStat), width=6, bg='yellow', relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+1, column=j)

            j = j+1

    #error messages for non-existent players and mismatching comparisons
    elif playerType(name) == 'n':
        response = messagebox.showerror('Invalid Player', "Player does not exist. Try re-typing the name of the player you are looking for. (Case sensitive)")
    else:
        response = messagebox.showerror('Invalid Player', "Cannot compare a batter with a pitcher.")

        
        

#exitCompare function
#function for exiting compare menu
def exitCompare():
    root.geometry('1250x130')
    global compareBtn, compareLabel, compareBar, searchBtn2, exitCompareBtn, searchBtn
    compareLabel.destroy()
    compareBar.destroy()
    searchBtn2.destroy()
    exitCompareBtn.destroy()

    compareBtn = Button(root, text='Compare to', width=9, font=('Calibri', fontSize), bg='#019FD8', command= compareTo)
    compareBtn.grid(row=1, column=3, columnspan=2, sticky=W)
    searchBtn = Button(root, text='Search', width=9, font=('Calibri', fontSize), bg='#33D801', command= lambda: search(searchBar.get()))
    searchBtn.grid(row=1, column=1, columnspan=2, sticky=W)
    
    Label(root, text= ' ', width=32, bg='white', font=('Calibri', fontSize)).grid(row=currentRow+2, column=0, padx=(10,0))
    j = 1
    for i in stats[1:]:
        Label(root, text= ' ', width=6, bg='white', font=('Calibri', fontSize)).grid(row=currentRow+2, column=j)
        j = j + 1
    search(searchBar.get())
    
    
    
#compare function   
#opens compare menu
def compareTo():
    root.geometry('1250x260')
    global compareBtn, compareLabel, compareBar, searchBtn2, exitCompareBtn, searchBtn
    compareBtn.destroy()
    searchBtn.destroy()
    compareLabel = Label(root, text="Enter a player to compare:", width=32, font=('Calibri', fontSize, 'italic'), bg='white', anchor=W)
    compareLabel.grid(row=2, column=0, columnspan=1, sticky=W, padx=(10,0))
    compareBar = Entry(root, width=32, font=('Calibri', fontSize), bg='light gray')
    compareBar.grid(row=3, column=0, columnspan=1, sticky=W, pady=5, padx=(10,0))
    searchBtn2 = Button(root, text='Compare', width=9, font=('Calibri', fontSize), bg='#019FD8', command= lambda: compare(compareBar.get()))
    searchBtn2.grid(row=3, column=1, columnspan=2, pady=5, sticky=W)
    exitCompareBtn = Button(root, text='X', width=4, font=('Calibri', fontSize), bg='#FF3333', command= exitCompare)
    exitCompareBtn.grid(row=3, column=3, columnspan=2, pady=5, sticky=W)
    
    
    
    
#batterSearch function
#batter search function
def batterSearch(name):
    global compareBtn
    try:
        root.geometry('1250x130')
        searchLabel.config(font=('Calibri', fontSize, 'italic'))
        searchBar.config(font=('Calibri', fontSize))
        searchBtn.config(font=('Calibri', fontSize))
        
        compareBtn.grid(row=1, column=3, columnspan=2, pady=5, sticky=W)
        
        Label(root, text= str(btc.loc[list(bi).index(name)].at['Name']), width=32, relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+1, column=0, padx=(10,0))
        Label(root, text= stats[0], width=32, relief='solid', font=('Calibri', fontSize, 'bold')).grid(row=currentRow, column=0, padx=(10,0))
        j = 1
        for i in stats[1:]:
            Label(root, text= stats[j], width=6, relief='solid', font=('Calibri', fontSize, 'bold')).grid(row=currentRow, column=j)
            Label(root, text= str(btc.loc[list(bi).index(name)].at[stats[j]]), width=6, relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+1, column=j)
            j = j + 1
    except ValueError:
         response = messagebox.showerror('Invalid Player', "Player does not exist. Try re-typing the name of the player you are looking for. (Case sensitive)")

            
            
#pitcherSearch function
#pitcher search function
def pitcherSearch(name):
    global compareBtn
    try:
        root.geometry('1250x130')
        searchLabel.config(font=('Calibri', fontSize))
        searchBar.config(font=('Calibri', fontSize))
        searchBtn.config(font=('Calibri', fontSize))
        
        compareBtn.grid(row=1, column=3, columnspan=2, pady=5, sticky=W)
        
        Label(root, text= pstats[0], width=32, relief='solid', font=('Calibri', fontSize, 'bold')).grid(row=currentRow, column=0, padx=(10,0))
        Label(root, text= str(ptc.loc[list(pi).index(name)].at['Name']), width=32, relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+1, column=0, padx=(10,0))
        j = 1
        for i in pstats[1:]:
            Label(root, text= pstats[j], width=6, relief='solid', font=('Calibri', fontSize, 'bold')).grid(row=currentRow, column=j)
            Label(root, text= str(ptc.loc[list(pi).index(name)].at[pstats[j]]), width=6, relief='solid', font=('Calibri', fontSize)).grid(row=currentRow+1, column=j)
            j = j + 1
        #block out extra batter stat block that pokes out 
        Label(root, text= ' ', width=6, bg='white', font=('Calibri', fontSize, 'bold')).grid(row=currentRow, column=j)
        Label(root, text= ' ', width=6, bg='white', font=('Calibri', fontSize)).grid(row=currentRow+1, column=j)
    except ValueError:
         response = messagebox.showerror('Invalid Player', "Player does not exist. Try re-typing the name of the player you are looking for. (Case sensitive)")


            
            
#start-up screen
searchLabel = Label(root, text="Enter a player's full name to search for:", width=32, font=('Calibri', 28, 'italic'), bg='white', anchor=W)
searchLabel.grid(row=0, column=0, columnspan=1, sticky=W, padx=(10,0))
searchBar = Entry(root, width=32, font=('Calibri', 28), bg='light gray')
searchBar.grid(row=1, column=0, columnspan=1, pady=5, sticky=W, padx=(10,0))
searchBtn = Button(root, text='Search', width=9, font=('Calibri', 28), bg='#33D801', command= lambda: search(searchBar.get()))
searchBtn.grid(row=1, column=1, columnspan=2, pady=5, sticky=W)
compareBtn = Button(root, text='Compare to', width=9, font=('Calibri', fontSize), bg='#019FD8', command= compareTo)

root.mainloop()
```
